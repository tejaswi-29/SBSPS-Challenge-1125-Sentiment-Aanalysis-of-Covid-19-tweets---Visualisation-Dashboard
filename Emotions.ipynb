```python
from ibm_watson import ToneAnalyzerV3
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
authenticator = IAMAuthenticator('paste the api key')
tone_analyzer = ToneAnalyzerV3(
    version='2017-09-21',
    authenticator=authenticator)
tone_analyzer.set_service_url('paste the url')
```

```python
from cloudant.client import Cloudant
from cloudant.error import CloudantException
from cloudant.result import Result, ResultByKey
```

I have manually created a database "tweets2" in cloudant db with 2 documents: "Lockdowns" and "Livefeed". Lockdowns is used below which stores the count of each emotion in the lockdown.


```python
client = Cloudant("paste username", "paste password", url='paste url')
client.connect()
print("connected to db!")
my_database = client['tweets2']
my_document = my_database['Lockdowns']
```

### Im letting the program sleep after every 3 tweets are analysed because Watson tone analyser throws an error of max retries exceeded and the service goes inactive.

```python
# lockdown1=[]
# emotions_score1 = {}
for i in range(len(l1)):
    print(i)
    if i%3==0:
        time.sleep(5)
    text = l1[i]
#     print(processed_tweets[i])
    tone_analysis = tone_analyzer.tone({'text': text},content_type='application/json').get_result()
    tone=json.dumps(tone_analysis, indent=2)    
    jsonParse = json.loads(tone)
    if "document_tone" in jsonParse.keys():
        if jsonParse["document_tone"]["tones"]!=[]:
            length = len(jsonParse["document_tone"]["tones"])
            k = jsonParse["document_tone"]['tones']
            for i in range(length):
                n = k[i]['tone_name']
                my_document['Lockdown1'][n]+=1
                my_document.save()
#                 if n in emotions_score1:
#                     emotions_score1[n] = emotions_score1[n]+k[0]['score']
#                 else:
#                     emotions_score1.update({n:k[0]['score']})     
        else:
            my_document['Lockdown1']['Neutral']+=1
            my_document.save()
    else:
        my_document['Lockdown1']['Neutral']+=1
        my_document.save()
```
